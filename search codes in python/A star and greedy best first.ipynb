### A* Search:


```python
graph = {'S' : [('A' , 1), ('B' , 4)],
         'A' : [('B' , 2) , ('C' , 5), ('G' , 12)],
         'B' : [('C' , 2)],
         'C' : [('G' , 3)],
         'G' : []}
H_Table = {'S' : 7,
           'A' : 6 ,
           'B' : 4,
           'C' : 2 ,
           'G' : 0 }
def path_fcost_finder(path): 
    g_cost = 0
    for node , cost in path:
        g_cost += cost
    last_node = path[-1][0]
    h_cost = H_Table[last_node]
    f_cost = h_cost + g_cost
    return f_cost , last_node
```

```python
def A_Star_Search(graph , start , goal):
    visited = []
    path = []
    queue = [[(start , 0)]]
    while queue:
        queue.sort(key = path_fcost_finder)
        path = queue.pop(0)
        node = path[-1][0]
        if node == goal:
            return path
        if node in visited:
            continue
        else:
            visited.append(node)
            adjacent = graph[node]
            for next_node,cost in adjacent:
                new_path = path.copy()
                new_path.append((next_node , cost))
                queue.append(new_path)
```

```python
A_Star_Search(graph , 'S' , 'G')
```

## Greedy Best-First Search:

```python
graph = {'S' : [('A' , 1), ('B' , 4)],
         'A' : [('B' , 2) , ('C' , 5), ('G' , 12)],
         'B' : [('C' , 2)],
         'C' : [('G' , 3)],
         'G' : []}
H_Table = {'S' : 7,
           'A' : 6 ,
           'B' : 4,
           'C' : 2 ,
           'G' : 0 }
def path_hcost_finder(path): 
    last_node = path[-1][0]
    h_cost = H_Table[last_node]
    return h_cost , last_node
```

```python
def Greedy_Best_Search(graph , start , goal):
    visited = []
    path = []
    queue = [[(start , 0)]]
    while queue:
        queue.sort(key = path_hcost_finder)
        path = queue.pop(0)
        node = path[-1][0]
        if node == goal:
            return path
        if node in visited:
            continue
        else:
            visited.append(node)
            adjacent = graph[node]
            for next_node,cost in adjacent:
                new_path = path.copy()
                new_path.append((next_node , cost))
                queue.append(new_path)
```

```python
Greedy_Best_Search(graph , 'S' , 'G')
```
